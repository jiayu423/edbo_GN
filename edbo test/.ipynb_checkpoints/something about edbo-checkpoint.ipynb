* we want to optimize for reaction yields, but we don't know the mapping between input and output
* the most straightforward way is to preform a grid search to find the optimal
* BO: after preform an experiment, the algo will tell you where to sample next 
* Build a surrogate model for the mapping

```python
def plot(bo): 
    mean = bo.obj.scaler.unstandardize(bo.model.predict(bo.obj.domain)) # gp posterior mean
#     mean = bo.model.predict(bo.obj.domain)
    std = np.sqrt(bo.model.variance(bo.obj.domain)) * bo.obj.scaler.std * 2  # gp posterior std
    
    target = bo.obj.target
#     current_res = (bo.obj.scaler.unstandardize_target(bo.obj.results, target)).to_numpy()
    current_res = (bo.obj.results).to_numpy()
    next_res = bo.obj.get_results(bo.proposed_experiments).to_numpy() # use computational model to simulate results
    
    # existing results and next experiments and their results
    plt.figure(figsize=(8, 8))
    
    plt.subplot(211)
    plt.plot(x, y, label='True function')
    plt.plot(x, mean, label='GP mean')
    plt.fill_between(x.flatten(), mean-std, mean+std, alpha=0.4)

    plt.subplot(212)
    plt.plot(x, y, label='True function')
    plt.scatter(current_res[:, 0], current_res[:, 1],  color='k', label='known experiments')
    plt.scatter(next_res[:, 0], next_res[:, 1],  color='r', label='next experiments')

    plt.legend()
    plt.show()
    
    return next_res[0][1][0]
```

```python
def f(X):
    return -np.sin(3*X) - X**2 + 0.7*X

def f_real(X, noise=0.): 
    return -np.sin(3*X) - X**2 + 0.7*X + np.random.randn(*X.shape) * noise

x = np.arange(-1, 2, 0.01).reshape(-1, 1)
y = f(x)

plt.plot(x, y)
plt.show()
```

```python
# need to convert the search space into a dataframe object
# x[0], x[-1] = lol1, lol2
X = pd.DataFrame(x, columns=['X'])
```

Initilize BO object and generate two experiment proposal using random method

```python
# init bo
target = 'f(x)' # give a name to the DataFrame column
acquisition_function='EI'
init_method='rand'
lengthscale_prior=[GammaPrior(1.2, 1.1), 0.2] # GP prior and initial value
noise_prior = [GammaPrior(1.2, 1.1), 0.2]
batch_size = 1 # number of exps proposed during each iteration
objective = f_real # known objective funtions, for testing purposes

bo = edbo.bro.BO(results_path='results.csv',
             domain=X, 
             target=target, 
             acquisition_function=acquisition_function,
             init_method=init_method,
             lengthscale_prior=lengthscale_prior, 
             noise_prior=noise_prior, 
             batch_size=batch_size, 
             fast_comp=False, # not using gpytorch to accelerate computation
             computational_objective=None) 
```

```python
# generate two initial experimental conditions
# using append will append simulated or indexed data to the existing dataset
init_proposal = bo.init_sample(append=True, seed=6)

# note that the gp is not trained in this stage
# init_exps = bo.obj.results_input().to_numpy() # bo.obj.results will return existing data
# plt.plot(x, y, label='True function')
# plt.plot(init_exps[:, 0], init_exps[:, 1], '.', color='r', label='Init experiments')
# plt.legend()
# plt.show()
```

Since we have the computational objective, lets simulate experimental design and inference and excutions

Lets try BO_express to see the difference
* bo_express pick gp prior based on design space dimension 
* so if we dont want to deal with csv (which bo_expess needs) we can just use classical bo instead

```python
# create experiment domains
vars_ = {'x': (-1, 2, 0.01)}
components, encoding = {}, {}
for var, (start, end, increment) in vars_.items():
    domain = np.arange(start, end, increment)
    components[var] = domain
    encoding[var] = 'numeric'
    
# create bo_express object
# note that noise is not specified, and lengthscale prior is not specified
bo = edbo.bro.BO_express(reaction_components=components,
                         encoding=encoding,
                         batch_size=2,
                         init_method='rand',
                         acquisition_function='EI',
                         target='f(x)', 
                         computational_objective=f)
```

```python
init_proposal = bo.init_sample(append=True, seed=5)
init_pro_unstd = x[init_proposal.index.to_numpy()]

# preform exp
init_result = bo.obj.get_results(pd.DataFrame(init_pro_unstd), append=False).to_numpy()

# add experiment to existing data


# note that the gp is not trained in this stage
plt.plot(x, y, label='True function')
plt.plot(init_result[:, 0], init_result[:, 1], '.', color='r', label='Init experiments')
plt.legend()
plt.show()
```

```python
a = np.array([1,2])
b = np.array([4,5,6])
c = np.array([7,8])
```

```python
def populate_design_space(arr_list, name_list): 
    total_len = np.prod([len(arr) for arr in arr_list])
    design_space = []
    prev_len = 1
    
    for arr in arr_list: 
        
        current_len = len(arr)
        new_dim = int(total_len / prev_len)
        arr = np.sort(np.tile(arr, int(total_len/current_len)).reshape(-1, new_dim), axis=1).reshape(-1, )
        prev_len *= current_len
        design_space.append(arr)
        
    design_dict = {}
    for arr, name in zip(design_space, name_list): 
        design_dict[name] = arr
        
    return design_dict
```

```python
populate_design_space([a, b, c], ['a', 'b', 'c'])
```
